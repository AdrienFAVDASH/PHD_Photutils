```python
from synthesizer.parametric import Stars, BlackHole
from synthesizer.parametric.galaxy import Galaxy
from synthesizer.parametric.morphology import Sersic2D
from synthesizer.parametric.sf_hist import Constant
from synthesizer.parametric.metal_dist import DeltaConstant
from synthesizer.emission_models.stellar.models import IntrinsicEmission
from synthesizer.grid import Grid
from synthesizer.instruments.filters import FilterCollection
from synthesizer.imaging.image import Image
from synthesizer.imaging.image_collection import ImageCollection
from astropy.cosmology import Planck18 as cosmo
import numpy as np
from unyt import Msun, arcsec, yr, kpc, angstrom, erg, s, Hz, cm
```

```python
grid = Grid(grid_name='test_grid.hdf5', grid_dir='/Users/adrien/synthesizer/tests/test_grid', spectra_to_read = None, read_lines = True, 
           new_lam = np.logspace(2, 5, 600) * angstrom)
morph = Sersic2D(r_eff = 5 * kpc, sersic_index = 1.5, ellipticity = 0.4, theta = 1.0, cosmo = None, redshift = None)
sfh = Constant(max_age = 100 * 10**6 * yr, min_age = 0 * yr, duration = None)
Zdist = DeltaConstant(metallicity = 0.01, log10metallicity = None)
```

```python
stars = Stars(log10ages = grid.log10ages, metallicities = grid.metallicity, initial_mass = 10**9 * Msun, morphology = morph, sfzh = None, 
              sf_hist = sfh, metal_dist = Zdist)
```

```python
galaxy = Galaxy(stars = stars, name = 'test_galaxy', black_holes = None, redshift = 5, centre = None)
```

```python
model = IntrinsicEmission(grid = grid, fesc = 0.1, label = 'intrinsic', fesc_ly_alpha = 1.0, escaped = None, reprocessed = None)
sed = galaxy.stars.get_spectra(emission_model = model, dust_curves = None, tau_v = None, fesc = None, covering_fraction = None, 
                               mask = None, verbose = True)
```

```python
galaxy.get_observed_spectra(cosmo)
filters = FilterCollection(filter_codes = ['JWST/NIRCam.F277W'], tophat_dict = None, generic_dict = None, filters = None, path = None, 
                           new_lam = grid.lam, fill_gaps = True, verbose = True)
```

```python
galaxy.get_photo_fnu(filters)
```

```python
img = galaxy.get_images_flux(resolution = 30/250 * kpc, fov = 30 * kpc, emission_model = model, img_type = 'smoothed', kernel = None, 
                             kernel_threshold = 1, nthreads = 1, limit_to = None)
img.plot_images(show = True, vmin = None, vmax = None, scaling_func = None, cmap = 'Greys_r')
```

```python
img['JWST/NIRCam.F277W'].arr
```

```python
noise = {f: 1e-34 * erg / s / cm**2 / Hz for f in img.keys()}
noise_img1 = img.apply_noise_from_stds(noise)
noise_img1.plot_images(show = True, vmin = None, vmax = None, scaling_func = None, cmap = 'Greys_r')
```

```python
noise_img1['JWST/NIRCam.F277W'].noise_arr
```

```python
SNR = {f: 10 for f in img.keys()}
depth = {f: 1e-33 * erg / s / cm**2 / Hz for f in img.keys()}
noise_img2 = img.apply_noise_from_snrs(snrs = SNR, depths = depth, aperture_radius = None)
noise_img2.plot_images(show = True, vmin = None, vmax = None, scaling_func = None, cmap = 'Greys_r')
```

```python
noise_img2['JWST/NIRCam.F277W'].noise_arr
```

```python
noise_img3 = img.apply_noise_arrays(noise_arrs = {f: np.full((250, 250), 1e-33) * erg / s / cm**2 / Hz for f in img.keys()})
noise_img3.plot_images(show = True, vmin = None, vmax = None, scaling_func = None, cmap = 'Greys_r')
```

```python
noise_img3['JWST/NIRCam.F277W'].noise_arr
```
